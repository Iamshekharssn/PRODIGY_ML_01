**Preprocessing**

```python
Y_train= train_data['SalePrice']

train_data= train_data.dropna(axis=1)
test_data= test_data.dropna(axis=1)

train_data_columns= train_data.columns
test_data_columns = test_data.columns

common_columns= train_data_columns.intersection(test_data_columns)

common_columns
```

```python
train_data = train_data[common_columns]
test_data= test_data[common_columns]
train_data
```

```python
train_data.isna().sum()
test_data
```

```python
categorical_columns = train_data.select_dtypes(include=['object']).columns.tolist()

  # Print the list of categorical columns
if categorical_columns:
    print("Categorical columns:")
    for col in categorical_columns:
        print(col)
else:
    print("No categorical columns found.")
```

```python
len(categorical_columns)
```

```python
# work with OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
train_data[categorical_columns] = encoder.fit_transform(train_data[categorical_columns])
train_data[categorical_columns]
```

```python
#To see the mappings of original categories to encoded values
for col, categories in zip(categorical_columns, encoder.categories_):
    print(f"\nColumn: {col}")
    for i, category in enumerate(categories):
        print(f"  {category}: {i}")
```

```python
test_data[categorical_columns] = encoder.fit_transform(test_data[categorical_columns])
test_data[categorical_columns]
```

```python
# Standardize the non-cathegorical values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
categorical_columns.append('Id')
columns_to_scale = [col for col in train_data.columns if col not in categorical_columns]
train_data[columns_to_scale] = scaler.fit_transform(train_data[columns_to_scale])
train_data
```

```python
test_data[columns_to_scale] = scaler.fit_transform(test_data[columns_to_scale])
test_data[columns_to_scale]
```

**Training & prediction**

```python
X_train = train_data
Y_train


linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)
```

```python
# predict

y_pred = linear_reg.predict(test_data)
```

```python
test_data['SalePrice']=y_pred
```

```python
# export to csv

test_data.reset_index(inplace=True)

test_data.set_index('Id', inplace=True)

test_data[['SalePrice']].to_csv('predicted_sale_prices.csv', header=['SalePrice'])
```
